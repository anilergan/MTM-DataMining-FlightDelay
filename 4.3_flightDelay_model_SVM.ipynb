```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv(r'G:\Other computers\My Laptop D14\Belgeler\CODING\Python.PROJECTS\MTM-DataMining-FlightDelay\manipulated_data\flightDelay_preprocessed_for_SVM.csv')

df = data.copy()
```

```python
df
```

```python
X, y = df.drop('DelayLabel', axis=1), df['DelayLabel']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
```

```python
# Standart ölçeklendirme (Standard Scaling)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# SVM modelini oluşturma ve eğitme
svm_model = SVC(kernel='linear', C=1.0)  # Lineer kernel kullanıldı, C parametresi belirlenebilir

svm_model.fit(X_train_scaled, y_train)
# Test seti üzerinde tahmin yapma
y_pred = svm_model.predict(X_test_scaled)

# Modelin performansını değerlendirme
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
