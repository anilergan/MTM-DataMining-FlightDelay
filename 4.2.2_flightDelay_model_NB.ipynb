```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.naive_bayes import MultinomialNB
```

```python
pd.set_option('display.max_columns', 35)
```

```python
data = pd.read_csv(r'G:\Other computers\My Laptop D14\Belgeler\CODING\Python.PROJECTS\MTM-DataMining-FlightDelay\manipulated_data\flightDelay_manipulated2.csv')
df = data.copy()
```

```python
df
```

```python
select_cols = ['Flight_Number_Marketing_Airline', 'DepTimeBlk',  'OriginAirport', 'DestAirport', 'Merged by Operating_Airline', 'DelayLabel']
df2 = df.loc[: ,select_cols]
df2
```

```python
df2_encoded = df2.copy()
```

```python
le = LabelEncoder()
df2_encoded['DepTimeBlk'] = le.fit_transform(df2_encoded['DepTimeBlk'])
df2_encoded['OriginAirport'] = le.fit_transform(df2_encoded['OriginAirport'])
df2_encoded['DestAirport'] = le.fit_transform(df2_encoded['DestAirport'])
df2_encoded['Merged by Operating_Airline'] = le.fit_transform(df2_encoded['Merged by Operating_Airline'])
df2_encoded['DelayLabel'] = le.fit_transform(df2_encoded['DelayLabel'])
```

```python
df2_encoded
```

```python
X = df2_encoded.drop('DelayLabel', axis=1).values
y = df2_encoded.iloc[:,-1].values
```

```python
print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=100, test_size=0.2)
```

```python
model_mnb = MultinomialNB()
model_mnb.fit(X_train, y_train)
```

```python
y_pred = model_mnb.predict(X_test)
```

```python
# Modelin başarı değerlendirmelerini gösterelim
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```
