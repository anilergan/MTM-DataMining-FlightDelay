DepDelay, DepDelayMinutes, DepDel15, DepartureDelayGroups sütunları yüksek korelasyon sebebiyle preprocess aşamasında birleştirilmişti.

**DepDelay**: Planlanan ve fiili kalkış saati arasındaki dakika farkı. Erken ayrılmalar negatif rakamlar gösteriyor.

**DepDelayMinutes**: Planlanan ve fiili kalkış saati arasındaki dakika farkı. Erken kalkışlar 0 olarak ayarlandı.

**DepDel15**: Kalkış Gecikme Göstergesi, 15 Dakika veya Daha Fazla (1=Evet)

**DepartureDelayGroups**: Kalkış Gecikmesi aralıkları, her (<-15'ten >180'e 15 dakika)



```python
target_cols = ['DepDelay', 'DepDelayMinutes', 'DepDel15', 'DepartureDelayGroups']

df[target_cols] = df['Merged by DepDelay'].str.split(' / ', expand=True)
df[target_cols] = df[target_cols].astype(float)
```
