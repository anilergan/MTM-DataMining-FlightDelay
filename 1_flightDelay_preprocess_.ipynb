```python
# value counts Ã¼zerinde n quantile hesaplamak iÃ§in bir algoritma yazalÄ±m.
def calc_quantile(dataframe_column):
       liste_vc = []
       for value in dataframe_column.value_counts().sort_values(ascending=True).values:
              for epoch in range(value):
                     liste_vc.append(value)

       quantile_low_index = int(((len(liste_vc) + 1) / 4))
       quantile_up_index = int((len(liste_vc) + 1) * 3/4)

       quantile_low = liste_vc[quantile_low_index]
       quantile_up = liste_vc[quantile_up_index]  

       return quantile_low, quantile_up
```

```python
q1, q3 = calc_quantile(df2['ArrTimeBlk'])
Iqr = q3-q1

threshold = 1.5

lower_bound = q1 - Iqr * threshold
# upper_bound = Q3 + 1.5 * IQR
print('q1: ', q1)
print('q3: ', q3)
print('Iqr: ', Iqr)
print('lower_bound: ', lower_bound)
```

#### Birliktelik Analizi:
**Contingency table (Capraz Tablo) Ã¼zerinden deÄerlendirilecek:**
- Chi-Square Testi
- CramÃ©r's V'
- Yule's Q ve Yule's Y

**AyrÄ±ca deÄerlendirilecek:**
- Apriori 
- FP-growth

```python
from scipy.stats import chi2_contingency
import numpy as np

# Chi-kare testi yapma
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-Kare DeÄeri: {chi2}")
print(f'P-deÄeri: {p:.10f}')
# print(f"P-deÄeri: %.5f" %p)

# CramÃ©r's V hesaplayÄ±n
num_obs = np.sum(contingency_table.values)
cramer_v = np.sqrt(chi2 / (num_obs * (min(contingency_table.shape) - 1)))
print(f"CramÃ©r's V: {cramer_v}")

# Yule's Q ve Y hesaplayÄ±n
observed_freq = contingency_table.values
total_obs = np.sum(observed_freq)
expected_freq = np.outer(np.sum(observed_freq, axis=1), np.sum(observed_freq, axis=0)) / total_obs

yule_q = np.sum((observed_freq - expected_freq) ** 2 / expected_freq)
yule_q /= total_obs
yule_y = (yule_q - total_obs) / (yule_q + total_obs)
print(f"Yule's Q: {yule_q}")
print(f"Yule's Y: {yule_y}")
```

**Chi-Square Ä°statistiÄi (chi2)**: Bu deÄer, gÃ¶zlemlenen ve beklenen frekanslar arasÄ±ndaki farkÄ±n bÃ¼yÃ¼klÃ¼ÄÃ¼nÃ¼ Ã¶lÃ§er. YÃ¼ksek bir chi-square istatistiÄi, gÃ¶zlemlenen ve beklenen frekanslar arasÄ±nda Ã¶nemli bir fark olduÄunu gÃ¶sterebilir.

**P-value (p)**: Bu deÄer, chi-square istatistiÄinin anlamlÄ±lÄ±ÄÄ±nÄ± Ã¶lÃ§er. P-value genellikle belirli bir anlamlÄ±lÄ±k dÃ¼zeyinde (Ã¶rneÄin, 0.05) karÅÄ±laÅtÄ±rÄ±lÄ±r. EÄer p-value belirlenen anlamlÄ±lÄ±k dÃ¼zeyinden kÃ¼Ã§Ã¼kse, H0 hipotezi (iki deÄiÅken arasÄ±nda baÄÄ±msÄ±zlÄ±k) reddedilir.

**CramÃ©r's V deÄeri:**, Ä°ki kategorik deÄiÅken arasÄ±ndaki iliÅkiyi Ã¶lÃ§en bir normalized Ã¶lÃ§Ã¼dÃ¼r. DeÄer 0 ile 1 arasÄ±nda deÄiÅir, 0 zayÄ±f iliÅkiyi, 1 gÃ¼Ã§lÃ¼ iliÅkiyi gÃ¶sterir. YÃ¼ksek bir CramÃ©r's V deÄeri, iki deÄiÅken arasÄ±nda gÃ¼Ã§lÃ¼ bir iliÅki olduÄunu gÃ¶sterir.

**Yule's Q Ä°statistiÄi (yule_q)**: Bu istatistik, iki kategorik deÄiÅken arasÄ±ndaki iliÅkiyi Ã¶lÃ§er ve -1 ile 1 arasÄ±nda bir deÄer alÄ±r. -1'e yaklaÅan deÄerler tam ters orantÄ±lÄ± bir iliÅkiyi, 1'e yaklaÅan deÄerler doÄrusal bir iliÅkiyi gÃ¶sterir. 0, baÄÄ±msÄ±zlÄ±ÄÄ± ifade eder.

**Yule's Y Ä°statistiÄi (yule_y)**: Yule's Y, -1 ile 1 arasÄ±nda bir deÄer alÄ±r. 1'e yaklaÅan deÄerler gÃ¼Ã§lÃ¼ bir iliÅkiyi, -1'e yaklaÅan deÄerler ters orantÄ±lÄ± bir iliÅkiyi gÃ¶sterir.

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

def observation_of_columns_association_with_cramer(data):
    np.seterr(divide='ignore', invalid='ignore')

    data_col_list = data.columns.tolist()
    df_association = pd.DataFrame(columns=['Col1', 'Col2', 'Cramer'])

    epoch = 1
    for col1 in data_col_list:
        for col2 in data_col_list:

            if col1 == col2: continue
            if epoch % 100 == 0: print(f'{epoch}/{len(data_col_list) * (len(data_col_list)-1)}')

            contingency_table = pd.crosstab(data[col1], data[col2])
            chi2, _, _, _ = chi2_contingency(contingency_table)
            num_obs = np.sum(contingency_table.values)
            cramer_v = np.sqrt(chi2 / (num_obs * (min(contingency_table.shape) - 1)))

            epoch += 1

            if np.isnan(cramer_v): continue

            df_association = pd.concat([df_association, pd.DataFrame([col1, col2, cramer_v], index=df_association.columns).T],
                                       ignore_index=True)
            
        data_col_list.remove(col1)

    return df_association.sort_values(by=['Cramer'], ascending=False)


# Kodunuzun Ã§alÄ±ÅtÄ±rÄ±lmasÄ±
df_association = observation_of_columns_association_with_cramer(df3.iloc[:int(len(df3)/5), :])
```

```python
def concatinate_cols_according_to_cramer(data, threshold):
    np.seterr(divide='ignore', invalid='ignore')

    data_col_list = data.columns.tolist()
    df_merged = pd.DataFrame()

    for col1 in data_col_list:
        selected_cols = [col1]
        col1_dtype = data[col1].dtype

        for col2 in data_col_list:
            if col1 == col2: continue
            col2_dtype = data[col2].dtype
        
            # if not(((col1_dtype == object) & (col2_dtype == object)) | ((col1_dtype == object) & (col2_dtype != object)) | ((col1_dtype != object) & (col2_dtype == object))): continue

            contingency_table = pd.crosstab(data[col1], data[col2])
            
            chi2, _, _, _ = chi2_contingency(contingency_table)

            num_obs = np.sum(contingency_table.values)
            cramer_v = np.sqrt(chi2 / (num_obs * (min(contingency_table.shape) - 1)))

            if cramer_v >= threshold:
                selected_cols.append(col2)


        len_selected_cols = len(selected_cols)       
        if len_selected_cols > 1:
            df_merged[f'Merged by {col1}'] = data[selected_cols].astype(str).agg(' / '.join, axis=1).to_frame(' / '.join(selected_cols))
            
            for item in selected_cols: 
                data_col_list.remove(item)
                data.drop(item, axis=1, inplace=True)
    
    return data, df_merged
            
        
df_org, df_merged = concatinate_cols_according_to_cramer(data=df3, threshold=0.92)
```
